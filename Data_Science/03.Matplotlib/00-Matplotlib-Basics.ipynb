```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0, 10)
y = x*2
```

```python
plt.plot(x,y);
plt.title('Plot Title')
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.xlim(0, 9)
plt.ylim(0, 20)

plt.savefig('my_first_plot.png')
```
