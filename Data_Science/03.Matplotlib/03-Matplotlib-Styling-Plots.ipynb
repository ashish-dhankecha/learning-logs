```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = np.arange(10)
```

```python
fig = plt.figure()
ax = fig.add_subplot([0,0,1,1])
ax.plot(a,a**2,label='square');
ax.plot(a,a,label='linear');
ax.legend(loc=9)
ax.set_xlim(0,10)
ax.set_ylim(0,90)
```

```python
x = np.linspace(0,11,10)
```

```python
fig = plt.figure()
ax = fig.add_subplot([0,0,1,1])
ax.plot(x,x,color='#8619BF',label="x vs x",lw=10);
ax.plot(x,x**2,color='green',label="x vs x**2",ls="-.");
ax.legend()
```

```python
fig = plt.figure()
ax = fig.add_subplot([0,0,1,1])
lines = ax.plot(x,x,color='#8619BF');
lines[0].set_dashes([5,1,10,1])
```

```python
fig = plt.figure()
ax = fig.add_subplot([0,0,1,1])

ax.plot(x,x+2,color='red',marker='s',ls="--",ms=10,
        markerfacecolor='yellow',markeredgecolor='green',markeredgewidth=4);
```
