```python
fig = plt.figure()
axes1= fig.add_axes([0,0,1,1])
axes1.plot(a,b);
axes1.set_title('power to 4')
axes1.set_xlabel('A')
axes1.set_ylabel('B')
axes1.set_xlim(0,10)
axes1.set_ylim(0,10000)

axes2 = fig.add_axes([1,0.75,0.25,0.25])
axes2.plot(a,b)
axes2.set_title('zoomed in')
axes2.set_xlabel('A')
axes2.set_ylabel('B')
axes2.set_xlim(1,2)
axes2.set_ylim(0,50)
```

```python
fig = plt.figure(
    figsize = (10,5),
    dpi = 200
)
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)
fig.savefig("new_figure.png",bbox_inches='tight')
```
