```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = np.linspace(0,10,11)
b = a**4
x = np.arange(10)
y=x*2
```

```python
fig,axis = plt.subplots(2,2)
fig.suptitle('Matplotlib Subplots')
axis[0][0].plot(x,y)
axis[0][0].set_title("First plot")
axis[1][1].plot(a,b)
axis[1][1].set_title("fourth plot")
plt.tight_layout()
# fig.subplots_adjust(hspace=0.3,wspace=0.3,top=0.9,bottom=0.1)
fig.savefig("new_subplots.jpg",bbox_inches='tight')
```
